```python
#pandas to access and manupulate data and data sets dataframes etc
#numpy for mathematical cal and matrix manupulation arrays etc
import numpy as np 
import pandas as pd 
#reading the csv files
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv('tmdb_5000_credits.csv') 
#.head() is showing the movies data
movies.head()
```

```python
credits.head()
```

```python
movies.shape
```

```python
#merging the movies and credits
#we are merging both on based on title which both data_sets have in common 
movies=movies.merge(credits,on='title')
movies.head()
```

```python
#rewriting the movies data_set into only columns which we wanted for further processing
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
#we want to remove all the unwanted data and just want the needed data like
#acion adventure comdey etc and remove uncessary data
movies['genres']
```

```python
#here its a '[{}]' its a string of list of dict we cant access them as we did list
#cause its a string
movies.iloc[0].genres
```

```python
# we use ast module to remove the string and make the data a list item for processing
import ast
```

```python
#this function returns a list of having only neede data like  action etc
#we are accessing i['name'] because dict is a element inside list its a list element

def convert(text):
    l=[] 
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
#applying the function to the genres and rewriting the existing genres 
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#function returning a list of top 3 members from the crew list for every movie
#in the data_set
def convert_2(text):
    l=[]
    counter=0 
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
            counter=counter+1 
        else:
            break 
    return l
        
```

```python
movies['cast']=movies['cast'].apply(convert_2)
```

```python
movies.head()
```

```python
def fetch_director(text):
    l=[] 
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            l.append(i['name']) 
            break 
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#we are defining this function so that we can remove spaces between words 
#ex-sai pranav this name belong to one person but having space between it makes it two
#different name and if other person have a name like sai or pranav it counts as
#as  a same name when its not so we need to replace the space between the words
def space_remover(arr):
    l=[]
    for i in arr:
        #i is the elements in a array we are replace all the " "with "" in the
        #each and every element in the array 
        #replace is a string method not array method 
        #we are replacing all the " " and storing it in a list and updating that with original
        #we do this for each and every row (movies) in data_set
        l.append(i.replace(" ","")) 
    return l
```

```python
movies['genres']=movies['genres'].apply(space_remover)
movies['keywords']=movies['keywords'].apply(space_remover) 
movies['cast']=movies['cast'].apply(space_remover)
movies['crew']=movies['crew'].apply(space_remover)
```

```python
movies.head()
```

```python
#here all the categories genres cast crew keywords are in list format 
#overview is in a string format but we want to convert all the categories in a 
#new category to process so we are applying lambda split function on the overview
#split() function returns a list of string seperated by the space
#lambda is a function can take multiple arguments and apply a single expression

movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#all the string is converted into a list of strings
movies['overview']
```

```python
#creating a clumn tags contains all the other columns data
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
#we dont need all the other anymore so creating a new dataframe just including what we needed
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
# we need all the tags data which is in list into a big text (string ) so that we can
#perform vectorization on the text 
def join_all(arr):
    #join function retuns a string by joining with a specifier in a iterable data type
    #ex=["sai","pranav","dasari"]
    #string=" ".join(ex)
    #sai pranav dasari
    #no need of for or anyother loops
    return " ".join(arr)
```

```python
new_df['tags']=new_df['tags'].apply(join_all)
```

```python
new_df.head()
```

```python
#we use countvectorizer to vectorize the text in tags 
from sklearn.feature_extraction.text import CountVectorizer
#it gathers the max_features or most repeated features 5000 features from the large text
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#here we are creating a array of values that shows us the count of each most repeated
#words in all the tags given
#it is a matrix of 5000 rows and 5000 columns

vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#we import cosine_similarity 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#we need to calculate the cosine distances between vectors not the euclid 
#distances the more the distance less the similarity
#this returns a array of cosine distances of one movies to the other 5000 movies
#5000 rows and 5000 colums of distances between every movis vector to movie
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
#final function to recommend the movies taking the index from the movie from user
#afer getting the index we need to get the maximum 5 values of similarity cosine
#of that movies thats why we sort in reverse order and get the first 5 with high similarites
#we use enumerate so that we dont lose the index while we sort acc to the similarittes values
#we sort acc to 2 element and lastly print the movies
def recommend(movie):
    index=new_df[new_df['title']==movie].index[0] 
    distances=similarity[index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```
